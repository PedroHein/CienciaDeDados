# **Análises**


## **Q1.** Qual a tendência de óbitos por mês/ano em acidentes de trânsito?


```python
#Criação de df, filtrando mes/ano e agrupando a quantidade de óbitos
df_obitos = pd.to_datetime(dfacidentes['Data do Óbito']).dt.strftime('%Y-%m').reset_index()
df_obitos = df_obitos.groupby(['Data do Óbito']).size().reset_index(name='Quantidade de Óbitos')
df_obitos.head()
```

```python
#Converter a data para date_time
df_obitos['Data do Óbito'] = pd.to_datetime(df_obitos['Data do Óbito'])

#Plotar gráfico
plt.figure(figsize=(15, 6))
plt.plot(df_obitos['Data do Óbito'], df_obitos['Quantidade de Óbitos'], marker='o', linestyle='-', c='blue',mfc='blue')
plt.axvline(pd.Timestamp('2020-04'), color='red', linestyle='--')
#Função para plotar as faixas vermelhas no gráfico
for year in range(2015, 2024):
    start_date = pd.Timestamp(f'{year}-12')
    end_date = pd.Timestamp(f'{year + 1}-02')
    plt.axvspan(start_date, end_date, color='red', alpha=0.2)
plt.title('Quantidade de óbitos por mês no estado de São Paulo')
plt.ylabel('Quantidade de óbitos')
plt.show()
```

###**Conclusão**
###Embora não consigamos concluir uma tendência em relação aos óbitos, podemos observar um padrão de queda no final e inicio de cada ano (faixas vermelhas). Além disso, observamos uma queda brusca no mês de abril de 2020, isto se deve a declaração de lockdown do governo do Estado de São Paulo.

## **Q2.** Qual a idade e sexo das pessoas que se envolveram em acidentes fatais?

```python
# Criação de novo df, filtrando o sexo e a idade das vítimas
dfacidentes['Idade da vítima'] = pd.to_numeric(dfacidentes['Idade da vítima'].replace('NAO DISPONIVEL', 0))
df_sexo_idade = dfacidentes.groupby(['Sexo', 'Idade da vítima']).size().reset_index(name='Quantidade de Óbitos')
df_sexo_idade = df_sexo_idade.pivot(index='Idade da vítima', columns='Sexo', values='Quantidade de Óbitos')
df_sexo_idade = df_sexo_idade.drop(df_sexo_idade.index[0])
df_sexo_idade = df_sexo_idade.fillna(0)
df_sexo_idade.head()
```

```python
#Plotagem do gráfico
plt.figure(figsize=(20,6))
sns.scatterplot(data=df_sexo_idade, x=df_sexo_idade.index, y="FEMININO", label="Feminino")
sns.scatterplot(data=df_sexo_idade, x=df_sexo_idade.index, y="MASCULINO", label="Masculino")
sns.scatterplot(data=df_sexo_idade, x=df_sexo_idade.index, y="NAO DISPONIVEL", label="Não disponível")
plt.title('Distribuição de óbitos por idade da vítima e gênero')
plt.xlabel('Idade da vítima')
plt.ylabel('Quantidade de óbitos')
plt.legend()
plt.grid(True)
plt.show()
```

###**Conclusão**
###A distribuição de óbitos por idade e gênero mostra que homens entre 20-22 anos representam a maioria dos casos, apresentando uma tendência de queda com o aumento da idade. Já para as mulheres, é possivel observar um leve aumento no número de casos com esta mesma faixa etária, seguindo com a mesma tendência de queda com o aumento da idade.

## **Q3.** Houve mudança no número de acidentes fatais entre o ano de 2022 e 2023?

```python
#Criação do df, filtrando os anos, agrupando os dados e selecionando os 5 primeiros municípios
df_obitos_municipio = dfacidentes[dfacidentes['Ano do Óbito'].isin([2022, 2023])]
df_obitos_municipio = df_obitos_municipio.groupby(['Ano do Óbito', 'Município']).size().reset_index(name='Quantidade de Óbitos')
df_obitos_municipio = df_obitos_municipio.pivot(index='Município', columns='Ano do Óbito', values='Quantidade de Óbitos')
df_obitos_municipio = df_obitos_municipio.sort_values(by=2023, ascending=False).head(5)

# Plotar o gráfico para 2022
plt.figure(figsize=(15, 6))
plt.suptitle('Quantidade de óbitos nos 5 municípios com mais casos')
plt.subplot(1, 2, 1)
df_obitos_municipio_2022 = df_obitos_municipio[2022].sort_values(ascending=False).head(5)
df_obitos_municipio_2022.plot(kind='bar')
plt.title('2022')
plt.ylabel('Quantidade de óbitos')
plt.grid(axis='y')

plt.subplot(1, 2, 2)
df_obitos_municipio_2023 = df_obitos_municipio[2023].sort_values(ascending=False).head(5)
df_obitos_municipio_2023.plot(kind='bar')
plt.title('2023')
plt.ylabel('Quantidade de óbitos')
plt.grid(axis='y')
plt.show()
```

###**Conclusão (ALTERAR)**
###Podemos observar que houve uma diferença leve de acidentes fatais entre os anos de 2022 e 2023, mas que foi suficiente para alterar a posição de cada município. Em 2022, temos
1.   São Paulo
2.   Guarulhos
3.   Campinas
4.   Sorocaba
5.   São Bernardo do Campo

###E em 2023, temos
1.   São Paulo
2.   Campinas
3.   Guarulhos
4.   São Bernardo do Campo
5.   Sorocaba








## **Q4.** Qual o meio de transporte e o tipo da via que mais ocorreram acidentes fatais?

```python
# Cria um dataframe com os dados de tipo de via e meio de transporte, desprezando os valores NAO DISPONIVEL
df_via_transporte = dfacidentes[(dfacidentes['Tipo de via'] != 'NAO DISPONIVEL') & (dfacidentes['Tipo do veículo da vítima'] != 'NAO DISPONIVEL')]
df_via_transporte = df_via_transporte.groupby(['Tipo do veículo da vítima', 'Tipo de via']).size().reset_index(name='Quantidade de Acidentes')
df_via_transporte = df_via_transporte.pivot(index='Tipo do veículo da vítima', columns='Tipo de via', values='Quantidade de Acidentes').fillna(0)
print(df_via_transporte)

# Plotagem do gráfico
df_via_transporte.plot(kind='barh', stacked=True, figsize=(10, 6))
plt.title('Quantidade de acidentes fatais por tipo de via e meio de transporte')
plt.xlabel('Quantidade de acidentes')
plt.ylabel('Tipo do veículo da vítima')
plt.show()
```

###**Conclusão**
###Podemos concluir que os meios de transporte e tipos de vias que mais acontecem acidentes fatais, são:
1.   Motocicletas em vias municipais
2.   Automóveis em rodovias
3.   Pedestres em vias municipais

(Não consideramos os dados "Não disponíveis" para melhor visualização dos dados)


## **Q5.** Qual o período e dia da semana que mais ocorreram acidentes fatais?


```python
#Criação do df, filtrando turnos e dias da semana, excluindo os não disponiveis e ordenando as linhas e colunas
df_turno_dia = dfacidentes.groupby(['Turno', 'Dia da Semana']).size().reset_index(name='Quantidade de Óbitos')
df_turno_dia = df_turno_dia.pivot(index='Dia da Semana', columns='Turno', values='Quantidade de Óbitos')
df_turno_dia = df_turno_dia.drop(columns=['NAO DISPONIVEL'])
df_turno_dia = df_turno_dia.reindex(index=['DOMINGO', 'SEGUNDA', 'TERÇA', 'QUARTA', 'QUINTA', 'SEXTA', 'SÁBADO'])
df_turno_dia = df_turno_dia.reindex(columns=['MANHA', 'TARDE', 'NOITE', 'MADRUGADA'])

#Plotagem do gráfico
plt.figure(figsize=(10, 6))
sns.heatmap(df_turno_dia, annot=True, fmt='d', cmap='Blues')
plt.title('Quantidade de óbitos por período e dia da semana')
plt.xlabel('Período do dia')
plt.ylabel('Dia da Semana')
plt.show()
```

###**Conclusão**
###Analisando este mapa de calor, observamos que os dias da semana e períodos com mais acidentes fatais, são:
1.   Sábado a noite
2.   Domingo a noite
3.   Sexta-feira a noite

###Já os dias da semana e períodos com menos acidentes fatais, são:
1.   Terça-feira de madrugada
2.   Quarta-feira de madrugada
3.   Quinta-feira de madrugada

(Não consideramos os dados "Não disponíveis" para melhor visualização dos dados)
